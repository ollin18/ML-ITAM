```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.svm import SVC
```

```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, show#, output_file #Descomentar para guardar el archivo .html
from bokeh.models.widgets import Panel, Tabs
from bokeh.models import Span
output_notebook()
```

```python
datos = pd.read_csv('andSVM.csv')
datos
```

```python
x = datos[['X1','X2']]
y = datos['y']
```

```python
clf = SVC(C=100, kernel='linear')
clf.fit(x,y)
```

```python
w1 = clf.coef_[0][0]
w2 = clf.coef_[0][1]
```

```python
w0 = clf.intercept_[0]
m = -w1/w2
```

```python
for vec in clf.support_vectors_:
    print(vec)
```

```python
espacio =  np.linspace(-0.5,3.2)
p = figure(width=800, plot_height=800, title=None)
p.circle(x.X1,x.X2, color=["red" if i==0 else "blue" for i in y],size=20, alpha=0.7)
p.line(espacio, -(w0 + w1*espacio)/w2)
for vec in clf.support_vectors_:
    p.line(espacio,m*espacio-vec[0]*m+vec[1], color= "gold")
show(p)
```

# CÃ­rculo

```python
x1 = np.array([np.random.uniform(-6,6) for i in range(10000)])
x2 = np.array([np.random.uniform(-6,6) for i in range(10000)])
```

```python
x1cuad = np.array([i**2 for i in x1])
x2cuad = np.array([i**2 for i in x2])
```

```python
X = np.array([[x1[i],x2[i],x1cuad[i],x2cuad[i]] for i in range(len(x1))])
```

```python
circulo = np.array([1 if x1cuad[i]+x2cuad[i] <= 9 else 0 for i in range(len(x1))])
```

```python
predic = SVC(kernel='linear', C=100)
```

```python
predic.fit(X,circulo)
```

```python
Y = predic.predict(X)
p = figure(width=800, plot_height=800, title=None)
p.circle(X[:,0],X[:,1],color=["red" if i==1 else "blue" for i in Y])
p.title.text = "CÃ­culo predicho"
p.xaxis.axis_label = "X1"
p.yaxis.axis_label = "X2"
show(p)  
```
